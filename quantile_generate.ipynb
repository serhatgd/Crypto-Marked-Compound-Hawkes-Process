```python
import pandas as pd
import numpy as np

def generate_full_csv():
    # This is the lambda calculated from your 'exponential_quantiles.csv' file
    # Calculation: lambda = -ln(1 - p) / Quantile
    estimated_lambda = 0.40637
    
    num_new_points = 1000000
    output_file = "exponential_1000000_quantiles.csv"
    
    print(f"Generating {num_new_points} quantiles using lambda={estimated_lambda}...")
    
    # Generate probabilities uniformly spaced between 0 and 1
    new_probs = np.linspace(0, 1, num_new_points + 2)[1:-1]
    
    # Calculate quantiles: Q(p) = -ln(1 - p) / lambda
    new_quantiles = -np.log(1 - new_probs) / estimated_lambda
    
    # Create DataFrame
    output_df = pd.DataFrame({
        'Probability': new_probs,
        'Quantile': new_quantiles
    })
    
    # Save
    output_df.to_csv(output_file, index=False)
    print(f"Done! Saved to {output_file}")

if __name__ == "__main__":
    generate_full_csv()
```
